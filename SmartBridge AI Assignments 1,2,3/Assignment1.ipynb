```python
import numpy as np
import pandas as pd

np.random.seed(10)
data = np.random.rand(4, 4)
df = pd.DataFrame(data, columns=['F-1', 'F-2', 'F-3', 'F-4'])
print(df)
```

```python
df = df.rename(columns={'F-1': 'Random value 1', 'F-2': 'Random value 2' , 'F-3': 'Random value 3', 'F-4':'Random vlaue 4'})
print(df)
```

```python
stats = df.describe()
print(stats)
```

```python
null_values = df.isnull()
column_data_types = df.dtypes
print("Null Values:")
print(null_values)
print("Column Data Types:")
print(column_data_types)
```

```python
disp_colmn = df.loc[:, ['Random value 2', 'Random value 3']]
print("Using loc method:")
print(disp_colmn)
```

```python
disp_colmn_iloc = df.iloc[:, [1, 2]]
print("Using iloc method:")
print(disp_colmn_iloc)
```
